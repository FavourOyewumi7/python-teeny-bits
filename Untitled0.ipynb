```python
import json
import pandas as pd
import os
table = pd.DataFrame(columns = ['Tranx ref', 'account', 'tranx amount', 'date'])
jsons = {}

```

```python
entries = os.scandir(r'/content/34')
places = []
for entry in entries:
  places.append(entry.name)

print(len(places))
print(places)
```

```python
print(places)
```

```python
val = []
temp = []

for place in places:
  print(place)
  with open(r'/content/34/{}'.format(place), 'r') as f:
    for line in f.readlines():
      place = line.split(' - ')
      temp.append(place[-1])
  print('load file complete ........')
  
```

```python
for line in temp:
    app = line.split(': ')
    val.append(app[-1])
    print('reference gotten, work begun')
    print('loading to json format')

    
```

```python
print(val)
```

```python
for va in val:
      if va[0] == '{':
        res = json.loads(va)
        jsons[res["reference"]] = res 
print('json loaded')
print('go ahead with your check with jsons[sample]')
```

```python
print(len(jsons))
```

```python
references = []
with open(r'/content/refs.txt', 'r') as fi:
  for l in fi.readlines():
    t = l.split("\n")
    references.append(t[0])
print(references)
print(len(references))
```

```python

for ref in references:
  if jsons[ref]:
    entry = {'Tranx ref':jsons[ref]["reference"], 'account':jsons[ref]["account"],'tranx amount':jsons[ref]["amount"], 'date':jsons[ref]["date"]}
  
  table = table.append(entry , ignore_index = True)
  
```
